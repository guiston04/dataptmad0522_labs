```python
query1 = """SELECT t.title_id AS TITLE_ID,
a.au_id AS AUTHOR_ID,
CAST((t.advance/ta.royaltyper/100) AS DECIMAL(10,2)) AS ADVANCE,
(t.price * s.qty * t.royalty / 100 * ta.royaltyper / 100) as SALES_ROYALTY
FROM authors as a INNER JOIN titleauthor as ta
on a.au_id = ta.au_id
INNER JOIN titles as t 
ON ta.title_id = t.title_id
INNER JOIN sales AS s
ON t.title_id = s.title_id"""
```

```python
query1_challenge1_step1 = pd.read_sql_query(query1, engine)
query1_challenge1_step1
```

```python
query2 = """SELECT t.title_id AS TITLE_ID, a.au_id AS AUTHOR_ID,
(t.advance/ta.royaltyper/100) as ADVANCE,
(t.price * s.qty * t.royalty / 100 * ta.royaltyper / 100) as SALES_ROYALTY
FROM authors as a INNER JOIN titleauthor as ta
on a.au_id = ta.au_id
INNER JOIN titles as t 
ON ta.title_id = t.title_id
INNER JOIN sales AS s
ON t.title_id = s.title_id
GROUP BY t.title_id, a.au_id"""
```

```python
query2_challenge1_step2 = pd.read_sql_query(query2, engine)
query2_challenge1_step2
```

```python
query3 = """SELECT a.au_id AS AUTHOR_ID,
(t.advance/ta.royaltyper/100) as ADVANCE,
(t.price * s.qty * t.royalty / 100 * ta.royaltyper / 100) as SALES_ROYALTY
FROM authors as a INNER JOIN titleauthor as ta
on a.au_id = ta.au_id
INNER JOIN titles as t 
ON ta.title_id = t.title_id
INNER JOIN sales AS s
ON t.title_id = s.title_id
GROUP BY a.au_id
ORDER BY SALES_ROYALTY DESC
LIMIT 3"""
```

```python
query3_challenge1_step3 = pd.read_sql_query(query3, engine)
query3_challenge1_step3
```
