***Challenge 1***

```python
query1 = """SELECT 
t.title_id AS TITLE_ID,
ta.au_id AS AUTHOR_ID,
(t.advance/(CAST(ta.royaltyper AS FLOAT)/100)) AS ADVANCE,
(t.price * s.qty * (CAST(t.royalty AS FLOAT) / 100) * (CAST(ta.royaltyper AS FLOAT) / 100))  as SALES_ROYALTY
FROM 
titles as t 
INNER JOIN titleauthor as ta
on t.title_id = ta.title_id
INNER JOIN sales AS s
ON t.title_id = s.title_id"""
```

```python
query1_challenge1_step1 = pd.read_sql_query(query1, engine)
query1_challenge1_step1
```

```python
query2 = """SELECT 
t.title_id AS TITLE_ID,
ta.au_id AS AUTHOR_ID,
SUM((t.price * s.qty * (CAST(t.royalty AS FLOAT) / 100) * (CAST(ta.royaltyper AS FLOAT) / 100)))  as SALES_ROYALTY
FROM 
titles as t 
INNER JOIN titleauthor as ta
on t.title_id = ta.title_id
INNER JOIN sales AS s
ON t.title_id = s.title_id
GROUP BY t.title_id, ta.au_id"""
```

```python
query2_challenge1_step2 = pd.read_sql_query(query2, engine)
query2_challenge1_step2
```

```python
query3 = """SELECT 
ta.au_id AS AUTHOR_ID,
SUM((t.price * s.qty * (CAST(t.royalty AS FLOAT) / 100) * (CAST(ta.royaltyper AS FLOAT) / 100)) + t.advance/(CAST(ta.royaltyper AS FLOAT)/100)) AS PROFITS
FROM 
titles as t 
INNER JOIN titleauthor as ta
on t.title_id = ta.title_id
INNER JOIN sales AS s
ON t.title_id = s.title_id
GROUP BY ta.au_id
ORDER BY PROFITS DESC
LIMIT 3"""
```

```python
query3_challenge1_step3 = pd.read_sql_query(query3, engine)
query3_challenge1_step3
```

CREATE TEMPORARY TABLE advance_royalty as 
SELECT 
t.title_id AS TITLE_ID, 
ta.au_id AS AUTHOR_ID, 
(t.advance/(CAST(ta.royaltyper AS FLOAT)/100)) AS ADVANCE,
(t.price * s.qty * (CAST(t.royalty AS FLOAT) / 100) * (CAST(ta.royaltyper AS FLOAT) / 100))  as SALES_ROYALTY
FROM titles as t 
INNER JOIN titleauthor as ta 
on t.title_id = ta.title_id 
INNER JOIN sales AS s 
ON t.title_id = s.title_id


SELECT AUTHOR_ID, 
SUM((ADVANCE + SALES_ROYALTY)) as PROFITS 
FROM advance_royalty 
GROUP BY AUTHOR_ID 
ORDER BY PROFITS 
DESC LIMIT 3
