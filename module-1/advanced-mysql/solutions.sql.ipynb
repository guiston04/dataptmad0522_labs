```python
query1 = """SELECT 
t.title_id AS TITLE_ID,
ta.au_id AS AUTHOR_ID,
(t.advance/(CAST(ta.royaltyper AS FLOAT)/100)) AS ADVANCE,
(t.price * s.qty * (CAST(t.royalty AS FLOAT) / 100) * (CAST(ta.royaltyper AS FLOAT) / 100))  as SALES_ROYALTY
FROM 
titles as t 
INNER JOIN titleauthor as ta
on t.title_id = ta.title_id
INNER JOIN sales AS s
ON t.title_id = s.title_id"""
```

```python
query1_challenge1_step1 = pd.read_sql_query(query1, engine)
query1_challenge1_step1
```

```python
query2 = """SELECT 
STEP_1.TITLE_ID AS TITLE_ID,
STEP_1.AUTHOR_ID AS AUTHOR_ID,
SUM(STEP_1.SALES_ROYALTY) AS ROYALTIES
FROM 
	(SELECT 
	t.title_id AS TITLE_ID,
	ta.au_id AS AUTHOR_ID,
	(t.advance/(CAST(ta.royaltyper AS FLOAT)/100)) AS ADVANCE,
	(t.price * s.qty * (CAST(t.royalty AS FLOAT) / 100) * (CAST(ta.royaltyper AS FLOAT) / 100))  as SALES_ROYALTY
	FROM 
	titles as t 
	INNER JOIN titleauthor as ta
	on t.title_id = ta.title_id
	INNER JOIN sales AS s
	ON t.title_id = s.title_id) AS STEP_1
GROUP BY TITLE_ID, AUTHOR_ID"""
```

```python
query2_challenge1_step2 = pd.read_sql_query(query2, engine)
query2_challenge1_step2
```

```python
query3 = """SELECT 
STEP_2.AUTHOR_ID AS AUTHOR_ID,
SUM(STEP_2.ADVANCE + STEP_2.ROYALTIES) AS PROFITS
FROM
	(SELECT 
	STEP_1.TITLE_ID AS TITLE_ID,
	STEP_1.AUTHOR_ID AS AUTHOR_ID,
	STEP_1.ADVANCE AS ADVANCE,
	SUM(STEP_1.SALES_ROYALTY) AS ROYALTIES
	FROM 
		(SELECT 
		t.title_id AS TITLE_ID,
		ta.au_id AS AUTHOR_ID,
		(t.advance*(CAST(ta.royaltyper AS FLOAT)/100)) AS ADVANCE,
		(t.price * s.qty * (CAST(t.royalty AS FLOAT) / 100) * (CAST(ta.royaltyper AS FLOAT) / 100))  as SALES_ROYALTY
		FROM 
		titles as t 
		INNER JOIN titleauthor as ta
		on t.title_id = ta.title_id
		INNER JOIN sales AS s
		ON t.title_id = s.title_id) AS STEP_1
	GROUP BY TITLE_ID, AUTHOR_ID) AS STEP_2
GROUP BY STEP_2.AUTHOR_ID
ORDER BY PROFITS DESC 
LIMIT 3"""
```

```python
query3_challenge1_step3 = pd.read_sql_query(query3, engine)
query3_challenge1_step3
```
