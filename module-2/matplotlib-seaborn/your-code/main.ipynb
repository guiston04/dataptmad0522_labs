#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

x = np.arange(0,100)
y = x**2
z = np.exp(x)

fig, axs = plt.subplots(ncols=1, nrows=2)
axs[0].plot(x, y)
axs[0].set_xlim(0, 100)
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].grid(False)
axs[1].plot(x, z, color = 'orange', linestyle = 'dashed')
axs[1].set_xlim(0, 100)
axs[1].set_xlabel('x')
axs[1].set_ylabel('z')
axs[1].grid(False)
axs[1].set_yscale('log')
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
x = np.arange(0,100)
y = x**2
z = np.exp(x)

fig, axs = plt.subplots(ncols=1, nrows=2)
axs[0].plot(x, y, label="x**2")
axs[0].set_xlim(0, 100)
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].grid(False)
axs[0].legend(loc='upper left')
axs[1].plot(x, z, color = 'orange', linestyle = 'dashed', label="e**x")
axs[1].set_xlim(0, 100)
axs[1].set_xlabel('x')
axs[1].set_ylabel('z')
axs[1].grid(False)
axs[1].set_yscale('log')
axs[1].legend(loc='upper left')
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
fitbit_month = fitbit[['Steps','Months_encoded']].copy()
fitbit_month = fitbit_month.groupby(by = ['Months_encoded'], as_index = False).mean()
fitbit_month.head()
```

```python
fitbit_work_weekend = fitbit[['Steps','Work_or_Weekend']].copy()
fitbit_work_weekend = fitbit_work_weekend.groupby(by = ['Work_or_Weekend'], as_index = False).mean()
fitbit_work_weekend.head()
```

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

x1 = fitbit_month['Months_encoded']
y1 = fitbit_month['Steps']

x2 = fitbit_work_weekend['Work_or_Weekend']
y2 = fitbit_work_weekend['Steps']

labels = ['Workday', 'Weekend']

fig, axs = plt.subplots(ncols=1, nrows=2, figsize = [12,4])
plt.subplots_adjust(hspace = 0.5)
plt.sca(axs[0])
plt.bar(x1, y1)
plt.xticks(x1)
axs[0].set_xlim(0, 12)
axs[0].set_xlabel('Month')
axs[0].set_ylabel('avg.steps')
axs[0].grid(False)

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

plt.sca(axs[1])
plt.bar(x2, y2)
plt.xticks(x2, labels)
axs[1].set_xlabel('workday or weekend')
axs[1].set_ylabel('avg.steps')
axs[1].grid(False)
plt.savefig('/Users/guillaumedieude/desktop/ironhack/curso/Los_Labs/dataptmad0522_labs/module-2/matplotlib-seaborn/figures/figures.png')
```

#### Plot the correlation matrix using seaborn.

```python
# your code here

# Compute the correlation matrix
corr = titanic.corr()
print(corr)

# Generate a mask for the upper triangle
#mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
